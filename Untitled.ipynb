```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn
```

```python
dataset = pd.read_csv("Titanic classification data.csv")
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.columns
```

```python
dataset.drop(columns=["Cabin","Ticket","Name"],inplace=True) 
```

```python
dataset.columns
```

```python
# Label encoding for Sex
dataset['Sex_Encoded'] = dataset['Sex'].map({'male': 0, 'female': 1})

# Label encoding for Embarked
dataset['Embarked_Encoded'] = dataset['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
dataset.head() 
```

```python
dataset.drop(columns=["Sex","Embarked"],inplace=True) 
```

```python
dataset.head() 
```

```python
dataset.isnull().sum() 
```

```python
dataset.fillna(dataset["Embarked_Encoded"].mean(),inplace=True)
```

```python
dataset.fillna(dataset["Age"].mean(),inplace=True)
```

```python
dataset.isnull().sum() 
```

# Sepration column x n y feauter engineering 

```python
x = dataset.drop(columns=["Survived"],errors="error")
y=dataset["Survived"]
```

# train , Test Split Model selection

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=45)
```

# Scale the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() 
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test) 
```

# logistic Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression() 
logistic_model.fit(x_train,y_train)
```

```python
logistic_model_pred = logistic_model.predict(x_test) 
logistic_model_pred 
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,logistic_model_pred))
print(classification_report(y_test,logistic_model_pred)) 
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree_model=DecisionTreeClassifier()
decision_tree_model.fit(x_train,y_train)
```

```python
decision_tree_model_pred = decision_tree_model.predict(x_test) 
decision_tree_model_pred 
```

```python
print(confusion_matrix(y_test,decision_tree_model_pred))
print(classification_report(y_test,decision_tree_model_pred)) 
```

# Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier()
random_forest.fit(x_train,y_train)
```

```python
random_forest_pred = random_forest.predict(x_test) 
random_forest_pred 
```

```python
print(confusion_matrix(y_test,random_forest_pred ))
print(classification_report(y_test,random_forest_pred))
```

# Support Vector Machine

```python
from sklearn.svm import SVC 
```

```python
svm_model = SVC()
svm_model = svm_model.fit(x_train,y_train) 
```

```python
svm_model_pred = svm_model.predict(x_test) 
svm_model_pred
```

# Evaluate and Prediction

```python
print(confusion_matrix(y_test,svm_model_pred))
print(classification_report(y_test,svm_model_pred))
```

# K Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN_model = KNeighborsClassifier()
KNN_model.fit(x_train,y_train) 
```

```python
KNN_model_pred = KNN_model.predict(x_test) 
KNN_model_pred 
```

# evalute 

```python
print(confusion_matrix(y_test,KNN_model_pred))
print(classification_report(y_test,KNN_model_pred))
```
